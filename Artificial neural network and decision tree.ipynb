```python
#Part A: Decision Tree Classifier
#Task A1: Load and Explore the Dataset
#Objective: Understand the structure of the data.
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns=iris.feature_names)
df["class"] = y

print(df.head())
```

```python
#Task A2: Train a Decision Tree
#Objective: Train a Decision Tree classifier using CART
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X, y)
```

```python
#Task A3: Visualise the Decision Tree
#Objective: Understand how the tree makes decisions
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 8))
plot_tree(dt,
feature_names=iris.feature_names,
class_names=iris.target_names,
filled=True)
plt.show()
```

```python
#Task A4: Regularisation and Overfitting
#Objective: Reduce overfitting by limiting tree depth.
dt_limited = DecisionTreeClassifier(max_depth=2, random_state=42)
dt_limited.fit(X, y)
```

```python
#probability
sample = [[5.0, 3.4, 1.5, 0.2]]
probs = dt_limited.predict_proba(sample)
for name, prob in zip(iris.target_names, probs[0]):
 print(f"{name}: {prob:.2f}")
```

```python
# Import necessary libraries
from sklearn.datasets import load_breast_cancer       # Dataset for classification
from sklearn.preprocessing import StandardScaler     # For normalizing features
from sklearn.neural_network import MLPClassifier     # ANN model
from sklearn.metrics import accuracy_score           # To check accuracy

# Load dataset
data = load_breast_cancer()                          # Load breast cancer dataset
X = data.data                                        # Feature values
y = data.target                                      # Target labels (0=malignant, 1=benign)

# Normalize the features
scaler = StandardScaler()                             # Create scaler object
X_scaled = scaler.fit_transform(X)                    # Scale features (mean=0, std=1)

# Create the ANN model
ann = MLPClassifier(
    hidden_layer_sizes=(10,),                         # One hidden layer with 10 neurons
    activation='relu',                                # ReLU activation function
    max_iter=1000,                                    # Maximum training iterations
    random_state=42                                   # Reproducible results
)

# Train the ANN model
ann.fit(X_scaled, y)                                  # Train the network on the scaled data

# Make predictions
y_pred_ann = ann.predict(X_scaled)                    # Predict on the same data

# Evaluate the model
accuracy = accuracy_score(y, y_pred_ann)             # Calculate accuracy
print("ANN Accuracy:", accuracy)                      # Print ANN accuracy
```

```python
# Import the pandas library so we can work with tables of data
import pandas as pd

# Load the wine dataset from the Downloads folder
# sep=';' tells pandas that the columns are separated by semicolons
df = pd.read_csv(r"C:\Users\hanaa\Downloads\wine+quality\winequality-red.csv", sep=';')

# Show the first 5 rows of the dataset
# This helps us check if the data loaded correctly
df.head()
```

```python
# Quality >= 6 → Good, else Poor
# Create a new column called "quality_label"
# We convert the numeric wine quality into two categories: Good or Poor
df['quality_label'] = df['quality'].apply(
    lambda x: 'Good' if x >= 6 else 'Poor'  # If quality >= 6 → Good, otherwise → Poor
)


# X will contain the input features used to train the model
# We remove the quality columns because they are the answers we want to predict
X = df.drop(['quality', 'quality_label'], axis=1)


# y will contain the target labels (what we want the model to predict)
# In this case it is the new column Good or Poor
y = df['quality_label']
```

```python
#Train a DecisionTreeClassifier using all features.
#Restrict the tree using:
#max depth = 4
# Import the Decision Tree classification algorithm
from sklearn.tree import DecisionTreeClassifier


# Create a Decision Tree model
# max_depth=4 limits the size of the tree to prevent overfitting
# random_state=42 ensures we get the same results every time we run the code
dt = DecisionTreeClassifier(max_depth=4, random_state=42)


# Train the model using the input features (X) and the labels (y)
# The model learns patterns between wine chemicals and quality
dt.fit(X, y)
```

```python
#Visualise the trained Decision Tree
# Import tools to draw the decision tree
from sklearn.tree import plot_tree

# Import matplotlib to display graphs and plots
import matplotlib.pyplot as plt


# Create a large figure so the tree is easier to see
plt.figure(figsize=(16,8))


# Draw the decision tree
plot_tree(
    dt,                     # the trained decision tree model
    feature_names=X.columns, # show the names of the wine chemical features
    class_names=dt.classes_, # show the class labels (Good / Poor)
    filled=True,             # color the nodes based on class
    rounded=True             # make the boxes rounded
)


# Display the tree plot
plt.show()
```

```python
# Calculate how important each chemical feature is for predicting wine quality
importances = pd.Series(
    dt.feature_importances_,  # importance values from the model
    index=X.columns           # attach the feature names
).sort_values(ascending=False)  # sort from most important to least


# Select the top 3 most important features
top_features = importances.head(3)


# Print the three most important wine characteristics
print(top_features)
```

```python
 # Create two example wines with chemical values
# Each list represents one wine sample
sample_wines = [
    [7.4,0.70,0.00,1.9,0.076,11,34,0.9978,3.51,0.56,9.4],
    [11.2,0.28,0.56,1.9,0.075,17,60,0.998,3.16,0.58,9.8]
]


# Use the trained model to predict if the wines are Good or Poor
predictions = dt.predict(sample_wines)


# Print the predictions for the two wines
print(predictions)
```

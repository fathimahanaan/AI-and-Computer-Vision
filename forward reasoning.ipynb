```python
workload = "high"
stress_level = "high"
if workload == "high" and stress_level == "high":
 decision = "Recommend rest"
else:
 decision = "No action required"
print("Agent decision:", decision)
# Question 1: What type of AI paradigm does this represent?
# Answer:
# This represents a Rule-Based AI system (Symbolic AI).
# The system follows predefined rules using if-else logic to make decisions.

# Question 2: Is this system data driven or knowledge driven?
# Answer:
# This system is knowledge-driven.
# The decision is based on predefined rules created by a human,
# rather than learning patterns from data.
```

```python
# Exercise 3: Knowledge Representation

# Store information about a person using a dictionary
person = {
    "name": "Alex",
    "sleep_hours": 4,
    "stress": "high"
}

# Apply logical rule to detect risk
if person["sleep_hours"] < 6 and person["stress"] == "high":
    print("High risk detected")
else:
    print("Risk level acceptable")

# Explanation:
# A dictionary is used to represent structured knowledge about a person.
# The system checks two conditions:
# 1. Sleep hours are less than 6
# 2. Stress level is high
# If both conditions are true, the system identifies a high risk situation.
# This demonstrates knowledge representation and logical reasoning in AI.
```

```python
# Extension: Modify the rule so sleep < 7 triggers a warning

if person["sleep_hours"] < 7:
    print("Warning: insufficient sleep")
else:
    print("Sleep level acceptable")
```

```python
facts = {
    "rain": True,
    "wet_ground": False,
    "slippery": False,
    "slow_down": False
}

# Rule 1: rain -> wet_ground
if facts["rain"]:
    facts["wet_ground"] = True

# Rule 2: wet_ground -> slippery
if facts["wet_ground"]:
    facts["slippery"] = True

# Rule 3: slippery -> slow_down
if facts["slippery"]:
    facts["slow_down"] = True

print("Facts:", facts)
print("Total facts:", sum(facts.values()))
```

```python
# Exercise 4: Forward Reasoning

# Store initial facts
facts = {
    "exam_soon": True,
    "sleep_deprived": True
}

# Apply logical rule to infer new knowledge
if facts["exam_soon"] and facts["sleep_deprived"]:
    facts["stress_high"] = True

# Print updated facts
print("Updated facts:", facts)

# Explanation:
# Forward reasoning (forward chaining) starts with known facts
# and applies rules to generate new facts.

# Known facts:
# - exam_soon = True
# - sleep_deprived = True

# Rule:
# If both are true, then stress_high becomes True.

# The system automatically infers new knowledge from existing facts.
```

```python
# Exercise 5: Diagnostic Challenge

# Model accuracy
accuracy = 0.92

# Dataset size
data_size = "small"

# Rule to detect possible overfitting
if accuracy > 0.9 and data_size == "small":
    conclusion = "Possible overfitting"
else:
    conclusion = "Model likely generalising"

# Print the conclusion
print(conclusion)

# Correct Answer: B

# B. The model may not generalise well.

# Explanation:
# The model has very high accuracy (92%) but the dataset is small.
# This suggests the model may have memorized the training data
# instead of learning general patterns.

# This problem is known as overfitting.
# Overfitted models perform well on training data but poorly on new data.
```

```python
# Import required libraries
import numpy as np
from sklearn.cluster import KMeans

# Define the 2D points
X = np.array([
    [0, 0],
    [0, 2],
    [10, 10],
    [10, 8]
])

# Fit KMeans model
kmeans = KMeans(n_clusters=2, random_state=0, n_init=10)
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Print number of unique labels
print(len(set(labels)))
```

```python
# Part 1: What is Deep Learning

# Deep Learning is a subset of machine learning that uses neural networks
# with multiple layers to learn patterns directly from data.

# Unlike traditional machine learning methods such as Decision Trees,
# deep learning does not require manual feature engineering.

# Instead, the model automatically learns useful representations
# from raw data through multiple layers.

# Each layer learns increasingly complex patterns.

# Example (image recognition):
# - Early layers detect edges
# - Middle layers detect shapes
# - Deeper layers detect objects such as faces or cars

# The term "deep" refers to the large number of layers in the neural network.
```

```python
# Part 2: Data Representation and Tensors

# Neural networks cannot process raw data like images, text, or audio directly.
# All input data must be converted into numerical form called tensors.

# Types of tensors:

# Scalar (0D tensor)
# A single number
example_scalar = 5

# Vector (1D tensor)
# A list of numbers
example_vector = [1, 2, 3, 4]

# Matrix (2D tensor)
# A table with rows and columns
example_matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

# Higher dimensional tensors
# Used for complex data such as images and videos.

# Example:
# In image datasets, each image is converted into a matrix of pixel values.

# In this lab:
# - Each row represents one image
# - Each column represents a pixel value
```

```python
# Part 3: The Artificial Neuron

# A neuron (also called a perceptron) is the basic building block
# of an artificial neural network.

# The neuron performs the following steps:

# 1. Receives multiple input values
inputs = [0.5, 0.8, 0.2]

# 2. Each input is multiplied by a weight
weights = [0.4, 0.7, 0.3]

# 3. The weighted inputs are summed and a bias is added
bias = 0.1

weighted_sum = (
    inputs[0]*weights[0] +
    inputs[1]*weights[1] +
    inputs[2]*weights[2] +
    bias
)

# 4. The result passes through an activation function
import numpy as np

# Example activation function: ReLU
output = max(0, weighted_sum)

print("Neuron output:", output)

# The activation function determines how strongly the neuron activates.
```

```python
# Activation Functions introduce non-linearity into neural networks.

# Common activation functions include:

# ReLU (Rectified Linear Unit)
# Returns 0 for negative values and keeps positive values.
relu_example = max(0, -3)

# Sigmoid
# Compresses values between 0 and 1
import math
sigmoid_example = 1 / (1 + math.exp(-1))

# Softmax
# Converts outputs into probabilities that sum to 1.
# Often used in classification problems.

# In this lab, ReLU and Softmax are used.
```

```python
# Step 1: Load the dataset
from tensorflow.keras.datasets import mnist

# MNIST dataset contains handwritten digits (0-9)
# It is already split into training and testing sets
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# Step 2: Prepare the data
# Each image is 28x28 pixels. We flatten them into vectors of length 784.
train_images = train_images.reshape((60000, 28 * 28))
test_images = test_images.reshape((10000, 28 * 28))

# Convert pixel values from integers 0-255 to floats 0-1 for better neural network performance
train_images = train_images.astype("float32") / 255
test_images = test_images.astype("float32") / 255
```

```python
# Step 3: Build the neural network
from tensorflow import keras
from tensorflow.keras import layers

# Sequential model: stack layers one after another
model = keras.Sequential([
    keras.Input(shape=(784,)),          # Input layer expects 784 features (pixels)
    layers.Dense(512, activation="relu"),  # Hidden layer with 512 neurons and ReLU activation
    layers.Dense(10, activation="softmax") # Output layer with 10 neurons (digits 0-9)
])
```

```python
# Step 4: Compile the model
model.compile(
    optimizer="rmsprop",                         # Optimizer adjusts weights during training
    loss="sparse_categorical_crossentropy",      # Loss function for integer labels (0-9)
    metrics=["accuracy"]                          # Track accuracy metric
)
```

```python
#Step 5: Train the model
model.fit(train_images, train_labels, epochs=5, batch_size=128)
#Step 6: Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test accuracy:", test_acc)
#Part 7: Making Predictions
predictions = model.predict(test_images)
 
```

```python
predicted_digit = predictions[0].argmax()  # Index with highest probability
print("Predicted digit for first test image:", predicted_digit)
```

```python
# Step 1: Load the Fashion-MNIST dataset
from tensorflow.keras.datasets import fashion_mnist
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
# Load dataset: 60,000 training images, 10,000 test images
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
train_images = train_images.reshape((60000, 28 * 28))
test_images = test_images.reshape((10000, 28 * 28))

# Convert pixel values from integers 0-255 to floats 0-1 for better neural network performance
train_images = train_images.astype("float32") / 255
test_images = test_images.astype("float32") / 255
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model = keras.Sequential([
keras.Input(shape=(784,)),
layers.Dense(512, activation="relu"),
layers.Dense(10, activation="softmax")
])
```

```python
# Step 6: Compile the model
model.compile(
    optimizer="adam",                         # Adam optimizer works well for DNNs
    loss="sparse_categorical_crossentropy",   # Suitable loss for integer labels (0-9)
    metrics=["accuracy"]                      # Track accuracy during training
)
```

```python
# Step 7: Train the model
history = model.fit(
    train_images, train_labels,
    epochs=5,             # Number of times the model sees the entire dataset
    batch_size=128,       # Number of samples per gradient update
    validation_split=0.1  # Optional: reserve 10% of training data for validation
)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test accuracy:", test_acc)
```

```python
# Step 9: Make predictions on at least 3 test images
predictions = model.predict(test_images[:3])  # Predict first 3 test images
```

```python
# Mapping of class indices to clothing names
class_names = ["T-shirt", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
# Print predicted classes for the first 3 test images
for i, pred in enumerate(predictions):
    predicted_class = np.argmax(pred)  # Class with highest probability
    print(f"Test image {i+1} predicted as: {class_names[predicted_class]}")
```
